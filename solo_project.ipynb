## Let's try to identify each subject and their dates.

```python
import pandas as pd

selected_columns_subject = ['wave', 'gender', 'age', 'race', 'importance_same_race', 'importance_same_religion', 'field',
                            'attractive_important', 'intelligence_important', 'funny_important', 'shared_interests_important',
                            'attractive', 'sincere', 'intelligence', 'funny', 'ambition']

selected_columns_partner = ['wave', 'gender_o', 'age_o', 'race_o', 'pref_o_attractive', 'pref_o_sincere', 'pref_o_intelligence', 'pref_o_funny',
                            'pref_o_ambitious', 'pref_o_shared_interests']

new_df = pd.concat([file[selected_columns_subject], file[selected_columns_partner]], axis=1)

new_df['subject_id'] = new_df[selected_columns_subject].apply(tuple, axis=1).rank(method='dense').astype(int)
new_df['partner_id'] = new_df[selected_columns_partner].apply(tuple, axis=1).rank(method='dense').astype(int)

subject_id_unique_count = new_df['subject_id'].nunique()
partner_id_unique_count = new_df['partner_id'].nunique()

print(f"Unique count in 'subject_id': {subject_id_unique_count}")
print(f"Unique count in 'partner_id': {partner_id_unique_count}")
```

## Some Attribute Information

 * gender: Gender of self  
 * age: Age of self  
 * age_o: Age of partner  
 * d_age: Difference in age  
 * race: Race of self  
 * race_o: Race of partner  
 * samerace: Whether the two persons have the same race or not.  
 * importance_same_race: How important is it that partner is of same race?  
 * importance_same_religion: How important is it that partner has same religion?  
 * field: Field of study   
 * expected_happy_with_sd_people: How happy do you expect to be with the people you meet during the speed-dating event?  
 * expected_num_interested_in_me: Out of the 20 people you will meet, how many do you expect will be interested in dating you?  
 * expected_num_matches: How many matches do you expect to get?  
 * like: Did you like your partner?  
 * guess_prob_liked: How likely do you think it is that your partner likes you?   
 * met: Have you met your partner before?  
 * decision: Decision at night of event.
 * decision_o: Decision of partner at night of event.  
 * match: Match (yes/no)

## Participant's scores of importance on different attributes change with age. While 'funny' and 'ambitious' increased, 'intelligent' and 'attractive' scores decreased.

```python
early20s_data = file[(file['age_group'] == '21-23') | (file['age_group'] == '<20')]
late20s_data = file[(file['age_group'] == '24-26') | (file['age_group'] == '27-29')]
older_data = file[(file['age_group'] == '>30')]

# CALCULATE MEANS EARLY
early20s_imp_funny = early20s_data['funny_important'].mean()
early20s_imp_ambition = early20s_data['ambition_important'].mean()
early20s_imp_intelligence = early20s_data['intelligence_important'].mean()
early20s_imp_sincere = early20s_data['sincere_important'].mean()
early20s_imp_attractive = early20s_data['attractive_important'].mean()

# CALCULATE MEANS LATE
late20s_imp_funny = late20s_data['funny_important'].mean()
late20s_imp_ambition = late20s_data['ambition_important'].mean()
late20s_imp_intelligence = late20s_data['intelligence_important'].mean()
late20s_imp_sincere = late20s_data['sincere_important'].mean()
late20s_imp_attractive = late20s_data['attractive_important'].mean()

# CALCULATE MEANS OLDER
older_imp_funny = older_data['funny_important'].mean()
older_imp_ambition = older_data['ambition_important'].mean()
older_imp_intelligence = older_data['intelligence_important'].mean()
older_imp_sincere = older_data['sincere_important'].mean()
older_imp_attractive = older_data['attractive_important'].mean()

# CREATE KEY AND VALUES LISTS
attributes = ['Funny', 'Ambitious', 'Intelligence', 'Sincere', 'Attractive']
importance_means_by_early20s = [early20s_imp_funny, early20s_imp_ambition, early20s_imp_intelligence, early20s_imp_sincere, early20s_imp_attractive]
importance_means_by_late20s = [late20s_imp_funny, late20s_imp_ambition, late20s_imp_intelligence, late20s_imp_sincere, late20s_imp_attractive]
importance_means_by_older = [older_imp_funny, older_imp_ambition, older_imp_intelligence, older_imp_sincere, older_imp_attractive]

# CREATE A DICTIONARY FROM THE TWO LISTS
my_dict_e = dict(zip(attributes, importance_means_by_early20s))
my_dict_e = {k: v for k, v in sorted(my_dict_e.items(), key=lambda item: item[1], reverse=True)}

my_dict_l = dict(zip(attributes, importance_means_by_late20s))
my_dict_l = {k: v for k, v in sorted(my_dict_l.items(), key=lambda item: item[1], reverse=True)}

my_dict_o = dict(zip(attributes, importance_means_by_older))
my_dict_o = {k: v for k, v in sorted(my_dict_o.items(), key=lambda item: item[1], reverse=True)}

# PLOT
bar_width = 0.2
# ?
x = np.arange(len(attributes))

plt.bar(x - bar_width, my_dict_e.values(), color='blue', width=bar_width, edgecolor='grey', label='Early 20s')
plt.bar(x, my_dict_l.values(), color='green', width=bar_width, edgecolor='grey', label='Late 20s')
plt.bar(x + bar_width, my_dict_o.values(), color='purple', width=bar_width, edgecolor='grey', label='Older')
plt.xlabel('Attributes')
plt.ylabel('Mean Importance Rating')
plt.title('Mean Importance Ratings for Different Attributes by Age')
plt.legend()
# this turns the numbers in the x axis into labels
plt.xticks(x, attributes)

plt.show()
```
